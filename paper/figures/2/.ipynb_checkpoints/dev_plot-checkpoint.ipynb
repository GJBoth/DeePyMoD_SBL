```python
import numpy as np
from phimal_utilities.analysis import load_tensorboard

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context='paper', style='white')
```

```python
df_deepmod = load_tensorboard('runs/deepmod/')
df_logprob = load_tensorboard('runs/logprob/')
```

```python
df_logprob.keys()
```

```python
coeff_keys = [key for key in df_logprob.keys() if key[:5]=='coeff']
lst_keys = [key for key in df_logprob.keys() if key[:5]=='lstsq']
mae_keys = [key for key in df_logprob.keys() if key[:3]=='mae']
```

```python
true_coeffs = np.zeros((1, 12))
true_coeffs[0, 2] = 0.1
true_coeffs[0, 5] = -1.0
```

```python
%config InlineBackend.figure_format = 'svg'

fig, axes = plt.subplots(figsize=(10, 2.5), nrows=1, ncols=2, constrained_layout=True) #5.5 is official width

# Left panel
ax = axes[0]
ax.plot(df_deepmod.index, np.mean(np.abs(df_deepmod[coeff_keys].to_numpy() - df_deepmod[lst_keys].to_numpy()), axis=1), label='MSE')
ax.plot(df_logprob.index, np.mean(np.abs(df_logprob[coeff_keys].to_numpy() - df_logprob[lst_keys].to_numpy()), axis=1), label='Log. prob.')
ax.legend()

ax.set_yscale('log')
ax.set_title('Distance to lst. sq. solution')
ax.set_xlabel('Epoch')
ax.text(-0.26, 1.01, 'A', transform=ax.transAxes, weight='bold')
ax.ticklabel_format(axis="x", style="sci", scilimits=(3, 3))


# Right panel
ax = axes[1]
ax.plot(df_logprob.index, df_deepmod['MSE_0'] +  df_deepmod['Regression_0'], label='MSE')
ax.plot(df_logprob.index, df_logprob['MSE_0'] +  df_logprob['Regression_0'], label='Log. prob.')

ax.legend()
ax.set_yscale('log')
ax.set_title('Total loss')
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.text(-0.26, 1.01, 'B', transform=ax.transAxes, weight='bold')
ax.ticklabel_format(axis="x", style="sci", scilimits=(3, 3))

fig.show()
plt.savefig('figure_2.pdf')
```
